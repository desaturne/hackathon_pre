```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image


image = cv2.imread("sitting.webp")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# STEP 4: Load pre-trained YOLOv8 pose model
model = YOLO('yolov8n-pose.pt')  # Use 'yolov8s-pose.pt' or 'yolov8m-pose.pt' for more accuracy

# STEP 5: Run inference
results = model.predict(source=image_rgb, save=False)

# STEP 6: Draw pose on image
res_plotted = results[0].plot()

# STEP 7: Display the image
plt.figure(figsize=(10, 10))
plt.imshow(res_plotted)
plt.axis('off')
plt.title("YOLOv8 Pose Estimation")
plt.show()
```
